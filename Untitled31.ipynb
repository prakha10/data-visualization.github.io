```python
df = pd.read_html('https://worldpopulationreview.com/country-rankings/cost-of-living-by-country')[0]
```

```python
df.info()
```

```python
#collect the country level information.
df1 = pd.read_html ('https://github.com/prakha10/Project-1/blob/main/Cost_of_Living_Index_2022.csv')[0]
```

```python
# This will allow us to use the latitide and longitude data alongside the living indices when ploting the map
df = df.merge(df1)

df.info()
```

```python
#Define details to be highlighted when hovering
cols = ['Rent Index',
       'Cost of Living Plus Rent Index', 'Groceries Index',
       'Restaurant Price Index', 'Local Purchasing Power Index']
```

```python
fig = px.scatter_mapbox(df,  lat="Latitude", lon="Longitude", 
                     hover_name="Country",size="Cost of Living Index",
                     color = 'Cost of Living Index',hover_data = cols
                     #projection="mollweide"
                     #projection="orthographic"
                     ,title = 'Cost of Living Index Across Countries in the World', zoom=0.3, 
                      height=600, color_continuous_scale = px.colors.qualitative.Set1)
fig.update_layout(height=600,width=1000, margin={"r":0,"t":0,"l":0,"b":0})
fig.update_layout(mapbox_style="open-street-map")
fig.show()
```

### The developed the country the higher the cost of living is it ?

 To answer this question, we examined the _cost of living index_ of countries such as the United States (70.13), Canada (70.22), South Africa (42.09), Uganda (33.5), Australia (77.75), and Jordan (49.19).
