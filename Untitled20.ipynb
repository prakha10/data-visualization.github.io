```python
import pandas as pa

df = pa.read_csv('https://raw.githubusercontent.com/prakha10/data-visualization.github.io/main/iris.csv')
head = ['SepalLength','SepalWidth','PedalLength','PedalWidth','Class'] 
df.columns = head
df.head().T
```

```python
df.SepalLength.mean()

```

```python
df.SepalLength.astype('int')
```

```python
df.SepalLength.agg(['mean','median','count', 'std'])
```

# Converting SepalLength to integer doesnt have any affect in the mean value.

#Gathering the mean, median, count and standard deviation of all columns when grouped by Class.

```python
df.groupby('Class').SepalLength.agg(['mean','median','count', 'std'])
```

```python
df.groupby('Class').SepalWidth.agg(['mean','median','count', 'std'])
```

```python
df.groupby('Class').PedalLength.agg(['mean','median','count', 'std'])
```

```python
df.groupby('Class').PedalWidth.agg(['mean','median','count', 'std'])
```
