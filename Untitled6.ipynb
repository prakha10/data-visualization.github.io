# Finding out the longest trip using duration minutes .

```python
%%bigquery --project datavids
SELECT *
FROM  `bigquery-public-data.austin_bikeshare.bikeshare_trips`
limit 7
```

# grouping the data set using the trip id and bike id . adding all the minutes .

```python
%%bigquery --project datavids df
SELECT 
  trip_id	, 
  bikeid,
  SUM(duration_minutes) as total_minutes
FROM  `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE bikeid is not null
GROUP BY trip_id ,bikeid
```
