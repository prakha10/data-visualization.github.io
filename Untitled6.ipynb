```python
%%bigquery --project datavids
SELECT 
  COUNT(*) as total_rows
FROM  `bigquery-public-data.austin_bikeshare.bikeshare_stations`
```

```python
%%bigquery --project datavids
SELECT 
  AVG(footprint_length)
FROM  `bigquery-public-data.austin_bikeshare.bikeshare_stations`
```

```python
%%bigquery --project datavids
SELECT 
  AVG(table.footprint_length) as average_footprint_length
FROM  `bigquery-public-data.austin_bikeshare.bikeshare_stations`as table
```

```python
%%bigquery --project datavids
SELECT *
FROM  `bigquery-public-data.austin_bikeshare.bikeshare_stations`
limit 7
```

```python
%%bigquery --project datavids df
SELECT 
  alternate_name, 
  address,
  SUM(footprint_length) as total_length
FROM  `bigquery-public-data.austin_bikeshare.bikeshare_stations`
WHERE alternate_name is not null
GROUP BY alternate_name,address
```
