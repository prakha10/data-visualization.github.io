#Examine the average trip on Austin bike share grouping by property_type column from the bikeshare_stations table.

```python
%%bigquery --project datavids

SELECT name, property_type
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
LIMIT 50
```

```python
%%bigquery --project datavids
WITH t as(
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE property_type)

SELECT *
FROM t as table1 JOIN t as table2 ON table1.property_type = table2.property_type
```

#Is average time on trips starting from 'parkland' longer than from 'sidewalks'?
 # lets look at the process .

```python
%%bigquery --project datavids
WITH t as(
SELECT *
FROM  `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE property_type= parkland)

SELECT table1.start_station_name, SUM(CASE WHEN table1.parkland < table2.sidewalk Then 1 ELSE 0 END) as average_time_longer 
FROM t as table1 JOIN t as table2 ON table1.property_type = table2.property_type
GROUP BY table1.start_station_name
```
