```python
import pandas as pd 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/prakha10/data-visualization.github.io/main/iris.csv',header = None)
head = ['SepalLength','SepalWidth','PedalLength','PedalWidth','Class'] 
#https://raw.githubusercontent.com/prakha10/data-visualization.github.io/main/iris.csv 
df.columns = head
df.head() 
```

#Geting the column names.

```python
df.columns
```

#Calling a column and finding it's mean.

```python
df.SepalLength
```

```python
df.SepalLength.mean()
```

#Finding the last row.

```python
df.iloc[149:]
```

#Melting the data keeping two columns of values and the type of flower.

```python
df_melt = df.melt(id_vars = 'Class',
                  value_vars = ['SepalLength','SepalWidth'],
                  var_name = ['colmuns'])
                  
df_melt.tail()
```

```python
df_melt.pivot(columns = 'colmuns', values = 'value')
```

#Selecting flowers that have pedal length greater than 5 and sepal length less than 7.

```python
df.query('PedalLength >5 and SepalLength < 7').head()
```

# Grouping the data by type of flower and compute mean and median.

```python
df.groupby(by = 'Class').agg(['mean','median'])
```
