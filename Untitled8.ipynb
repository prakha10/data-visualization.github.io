```python
%%bigquery --project datavids

SELECT MAX(start_station_name) as max_start_station_name
FROM  `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

#For the second most popular starting station.

```python
%%bigquery --project datavids
SELECT MAX(start_station_name) as second_most_start_station_name
FROM  `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name != (SELECT MAX(start_station_name) as max_start_station_name FROM  `bigquery-public-data.austin_bikeshare.bikeshare_trips`)
```

# For trips lasted over an hour and were a round trip (started and stopped at the same station)

```python
%%bigquery --project datavids

SELECT start_station_name, COUNTIF(start_station_name = end_station_name and duration_minutes > 60 ) as  max_round_trips_over_onehour
FROM  `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name,duration_minutes
Order BY max_round_trips_over_onehour DESC
```
